```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nacd_synth = pd.read_csv('../results/NACD_synthetic.csv')
support_synth = pd.read_csv('../results/SUPPORT_synthetic.csv')
support_cens = pd.read_csv('../results/SUPPORT_synth_censoring.csv')
nacd_cens = pd.read_csv('../results/NACD_synth_censoring.csv')
```

```python
# Note that identification of t and shared factors does not apply to the synthetic censoring models

support_cens['N_distinct'] = ([4] * 7 + [8] * 7 + [12] * 7 + [16] * 7) * 3
support_cens['N_shared'] = 20 - support_cens['N_distinct']

nacd_cens['N_distinct'] = ([4] * 7 + [8] * 7 + [12] * 7 + [16] * 7) * 3
nacd_cens['N_shared'] = 20 - nacd_cens['N_distinct']
```

```python
nacd_cens.groupby(['model', 'ld'])[
    ['e_auc', 'y_ci', 'e_in', 'e_out', 'e_in_max', 'e_out_max']].agg(
    ['mean', 'std']).round(2)
```

```python
import seaborn as sns
sns.set()
```

```python
adf = pd.concat(
    [nacd_synth, nacd_cens, support_cens],
    keys=['nacd_synth', 'support_synth', 'nacd_cens', 'support_cens'],
    names=['ds', 'idx'],
    axis=0)

adf = adf[(adf['model'] == 'DNMC')]

adf = adf.reset_index().set_index(
    ['ds', 'random_state', 'model', 'ld'])[
    [
        'e_in_max', 'e_out_max', 'e_in', 'e_out',
        't_in_max', 't_out_max', 't_in', 't_out',
        'shared_in_max', 'shared_out_max', 'shared_in', 'shared_out'
    ]].melt(
    ignore_index=False
).reset_index()
```

```python
adf['Features'] = adf['variable'].str.split('_').str[0].apply(lambda x: {
    'e': 'Susceptibility',
    't': 'Timing',
    'shared': 'Shared'
}[x])
adf['view'] = adf['variable'].str.split('_').str[1].apply(
    lambda x: {'in': 'True Features', 'out': 'Other Features'}[x]
)
adf['max'] = adf['variable'].str.split('_').apply(lambda x: len(x) > 2)
```

```python
adf
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 6))

sns.boxplot(
    data=adf[adf['max'] & (adf['Features'] == 'Susceptibility')], ax=ax[0],
    x='ld', y='value', hue='view',
    showfliers=False
)
sns.boxplot(
    data=adf[adf['max'] & (adf['Features'] == 'Timing')], ax=ax[1],
    x='ld', y='value', hue='view',
    showfliers=False
)
sns.boxplot(
    data=adf[adf['max'] & (adf['Features'] == 'Shared')], ax=ax[2],
    x='ld', y='value', hue='view',
    showfliers=False
)
# sns.catplot(
#     data=adf[~adf['max']], ax=ax,
#     x='ld', y='value', hue='view', col='Features',
#     kind='box', showfliers=False, sharey=False
# )
ax[0].set_title('Identification of features\n affecting susceptibility', fontsize=16)
ax[1].set_title('Identification of features\n affecting timing', fontsize=16)
ax[2].set_title('Identification of features\n affecting both', fontsize=16)

ax[0].set_ylabel('Mean coefficient magnitude', fontsize=16)
ax[1].set_ylabel('')
ax[2].set_ylabel('')

ax[0].set_xlabel('$\lambda_d$ magnitude', fontsize=16)
ax[1].set_xlabel('$\lambda_d$ magnitude', fontsize=16)
ax[2].set_xlabel('$\lambda_d$ magnitude', fontsize=16)

ax[0].set_xticklabels(['Weaker', 'Stronger'], fontsize=16)
ax[1].set_xticklabels(['Weaker', 'Stronger'], fontsize=16)
ax[2].set_xticklabels(['Weaker', 'Stronger'], fontsize=16)

ax[0].legend(loc='upper right', framealpha=.5, fontsize=14)
ax[1].legend(loc='upper right', framealpha=.5, fontsize=14)
ax[2].legend(loc='upper right', framealpha=.5, fontsize=14)

plt.tight_layout()
plt.savefig('../results/identification.pdf')
plt.show()
```
